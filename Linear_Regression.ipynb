# **1. Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# **2. Dataset**

```python
dataset = pd.read_csv('Salary.csv')

print(dataset.head())
```

# **3. Visualizing the Data**

```python
plt.plot(dataset['Salary'], dataset['Years_Experience'], marker='o', linestyle='-')
plt.xlabel('Years_Experience')
plt.ylabel('Salary')
plt.title('Years_Experience vs Salary')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# **4. Preparing Data for Modelling**

```python
X = dataset[['Years_Experience']]
Y = dataset[['Salary']]
```

# **5. Fitting the Linear Regression Model**

```python
model = LinearRegression()
model.fit(X,Y)
```

# **6. Getting Model Coefficients**

```python
slope = model.coef_[0]

intercept = model.intercept_

print('slope:',slope)
print('intercept:',intercept)
```

# **7. Visualizing the Linear Regression Line**

```python
plt.scatter(X,Y)
plt.plot(X,model.predict(X), color='yellow')
plt.xlabel('Years_Experience (X)')
plt.ylabel('Salary (Y)')
plt.title('Linear Regression')
plt.show()
```

# **8. Making Predictions**

```python
input_value = np.array([[7]])
predicted_Y = model.predict(input_value)

print(predicted_Y)
```
